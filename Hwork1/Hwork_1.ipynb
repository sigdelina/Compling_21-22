Я хотела бы в первую очередь проанализировать, какие теги в тексте встречаются.


*   Во-первых, есть теги, которые обозначаются открытием и закрытием: <><\>
*   Во-вторых, есть теги, которые могут только открываться и не закрытваться: <> (типа тега <br>, который не закрывается)

Так как в условии не указано, что это необходимо сделать за один этап, мы разделим этот блок для того, чтобы результат был чуть более независимым: очистим текст от сложных тегов, а после - удалим оставшиеся одиночные.

На самом деле, с помощью того алгоритма, который уже сделан, в данном случае нет необходимости проверять существование одиночных тегов, так как они все удалены, однако в случае появления каких-то сложных одинчных тегов, они могут не удалится и тогда нам потребуется удаление их. Более того, сделать удаление тегов можно еще проще (в одну строчку, я указала это ниже)


````python
text_no_tags = ''
````

````python
def make_list_of_tags(text_string):
   
    """
    выбираем из строки все блоки, где может содержаться тег с его метаданными
    и объединяем их в группы, а после - 
    составляем список, который содержал бы все уникальные теги
    """
    
    finding = re.findall(r'(<.*?>).*?(<.*?>)', text_string, re.DOTALL)

    list_for_tags = []

    for find in finding:
        if find[0] not in list_for_tags:
            list_for_tags.append(find[0])
        if find[1] not in list_for_tags:
            list_for_tags.append(find[1])
    
    return sorted(list_for_tags)
````

Перед тем, как я начала удалять теги, я обратила внимание, что, несмотря на то, что нашлись все необходимые теги, при удалении с помощью регулярных выражений нашлись не все.



```
<a href="https://www.google.com/analytics/home/?hl=en" title="Шикарный сбор и анализ статистики">
```

Этот тег из-за того, что у него есть специальные символы читается регуляками неверно и поэтому не удаляется с помошью них.

Исправим это с помощью функции адаптирования, чтобы это было более универсальным.



````python
def adaptate_tags_removing(tag):

    """
    некоторые теги из-за специальных символов могут быть не удалены
    """

    list_symb = ['/', '\?']

    for el in list_symb:
        tag = re.sub(el, '\\%s'%(el), tag)
    
    return tag
````

````python
def control_spaces(clean_text):

    """
    корректируем все, что связано с пробелами
    """

    # контролируем, чтобы было не больше 1 пробела
    clean_text = re.sub(r'\s\s+', ' ', clean_text)

    # чтобы перед знаком препинания не было пробела
    all_zn_prep = [',', '.', ':', '!', '?', ';']


    for zn in all_zn_prep:
        clean_text = re.sub(r'\s\%s'%(zn), '%s'%(zn), clean_text)

        # контролируем, чтобы текст не слипался со знаками препинания
        # так как даже после сокращений типа "т. д."
        # и не очень понятно, что с ними делать, поэтому посмотрим на концы предложений
        clean_text = re.sub(r'\%s([А-ЯЁ])'%(zn), r'%s \1'%(zn), clean_text)

    return clean_text
````

````python
def remove_tags(text):

    """
    используем список тегов, полученных регулярным выражением
    и удалим их с помощью регулярок из текста
    """
    
    tags_list = make_list_of_tags(text)

    clean_text = ''
    
    for tag in tags_list:
        text = re.sub(adaptate_tags_removing(tag), ' ', text)
    
#     # удаляем все, что могло остаться (тут нам это не нужно, но вдруг)
#     text = re.sub(r'(<.*?>', ' ', text)
    
#     text = re.sub(r'\n', '', text)
    # корректируем пробелы
    clean_text = control_spaces(text)
    return clean_text
````

На самом деле, весь этот алгоритм можно было бы упростить и свести к тому, чтобы мы удаляли просто все "<.*?>":  text = re.sub(r'(<.*?>', ' ', text) этого было бы достаточно, но мне казалось, что достаточно важным является то, что внутри тегов может быть текст и с помощью алгоритма, который это видит и выделяет группы в начале и конце, можно этот текст выделить.

Кроме того, можно было бы удалить значения \n, сделать это достаточно просто можно с помощью text = re.sub(r'\n', '', text), однако я побоялась убирать их, так как в задании четко не было сказано, что нужно убрать кроме лишних пробелов еще и операторы переноса строк. Мне кажется, что они все-таки достаточно важные, поэтому я решила их оставить и закомментировать это. Тем не менее, я понимаю, что для качественной работы стоит удалить и их, но в настоящем задании это не мешает работе, которая выполнялась, поэтому я пришла к выводу, показать, что это нужно сделать в том числе (закомментировав в коде), однако, за счет того, что не очень очевидно, нужно ли их удалять в рамках задания, я оставила переносы.

Так как список словоформ одниаков для случаев с дизамбигуацией и без нее мы будет оперировать 3 списками для того, чтобы сравнить леммы. Мне кажется, сравнивать именно леммы в данном случае достаточно хорошая мысль, так как мы можем +- четко проследить, в каком случае будет различие между "разметкой".

````python
# создадим счетчик индекса для приведения контекста
ind = 0

for lem_no_dis, lem_dis, words in zip(list_lemmas_no_dis,
                                      list_lemmas_dis, list_words_no_dis):
    
    context = ''
    if lem_no_dis != lem_dis:
        print('Словоформа:', words) 
        
        # собираем контексты
        if ind < 10:
            context = ' '.join(w for w in list_words_no_dis[:ind+10])
        elif ind + 10 > len(list_words_no_dis):
            context = ' '.join(w for w in list_words_no_dis[ind-10:])
        else:
            context = ' '.join(w for w in list_words_no_dis[ind-10:ind+10])
        
        print('Контекст: ', context)
        print('Лемма без дизамбигуации: ', lem_no_dis)
        print('Лемма с дизамбигуацией: ', lem_dis)
        print('___\n')
    
    ind += 1
````

Мы видим достаточно интересные результаты.

* Иногда результаты с дизамбигуацией и без нее нетривиальные: говорящим --> говорящий/говорить. Результат без дизамбигуации сводит причастие к глаголу, а с дизамбигуацией к начальной форме присчастия. Я думаю, что правильно было бы говорить о форме причастия, а не о форме глагола, так как это слишком жесткий подход, игнорирующий признаки прилагательного.

* Интересен вариант с "теплее". Без дизамбигуации это - прилагательное, а с ней - наречие. Мне кажется, в данном случае выигрывает результат разбора без дизамбигуации, так как вариант прилагательного ближе. Аналогично с "больше": без дизамбигауции: "больше", с ней - "много". Кажется, что "больше" подходит лучше, так как это ближе к прилагательному "большой". Однако в данном случае, как будет оба варианта не самые лучшие.

* "Может" --> может/мочь. Тут вариант мочь оказывается более верным, так как это не вводное слово, а глагол. Здесь лучше оказывается разметка без дизамбигуации.
